### The Prophet Forecasting Model

The Prophet uses a decomposable time series model with three main model components: trend, seasonality, and holidays. <br>

They are combined in the following equation:<br>
y(t)= g(t) + s(t) + h(t) + εt<br>
g(t): piecewise linear or logistic growth curve for modeling non-periodic changes in time series<br>
s(t): periodic changes (e.g. weekly/yearly seasonality)<br>
h(t): effects of holidays (user provided) with irregular schedules<br>
εt: error term accounts for any unusual changes not accommodated by the model<br>

Using time as a regressor, Prophet is trying to fit several linear and non linear functions of time as components. <br>
Modeling seasonality as an additive component is the same approach taken by exponential smoothing in Holt-Winters technique.<br>
Prophet is framing the forecasting problem as a curve-fitting exercise rather than looking explicitly at the time based <br>dependence of each observation within a time series.

### Prophet functionality for Time Series Cross Validation

Prophet includes functionality for time series cross validation to `measure forecast error` using historical data. <br>
This is done by selecting cutoff points in the history, and for each of them fitting the model using data only up <br>
to that cutoff point. We can then compare the forecasted values to the actual values<br>

Understanding fbProphet cross validation terms (Initial, Horizon and period)<br>
`Initial` – Initial training period. Occurs just at the start<br>
`Horizon` – Access prediction performance(`test`)<br>
`Period` – Interval between testing<br>

<a id='paraGrid'></a>
### Model hyper-parameter tunning 

With ParameterGrid<br>
Specify the params you want to hyper-tune by creating a dictionary, stating the exact param name <br>
and the corresponding list of values you want to try out inside a list<br>
Then, create a parameterGrid object using the previously created dict as an argument<br>
Lastly, use a for-loop to run through the items inside the grid object<br>

In this part, we will fit a model after performing hyper-parameter tunning

```python
train = combined_partial.iloc[:568]
test = combined_partial.iloc[568:]
```

```python
# https://www.kaggle.com/manovirat/timeseries-using-prophet-hyperparameter-tuning
# 
params_grid = {'changepoint_prior_scale':[0.75,1],
              'seasonality_prior_scale':[10,15],
              'n_changepoints' : [20,25],
             'interval_width':[0.85,0.95]
              }
grid = ParameterGrid(params_grid)
cnt = 0
for p in grid:
    cnt = cnt+1

print('Total Possible Models',cnt)
```

m = Prophet(
changepoint_prior_scale= 0.075,
holidays_prior_scale= 0,
n_changepoints= 10,
seasonality_prior_scale= 5,
seasonality_mode='additive'
)
m.fit(train)

```python
def get_model_params():
    """ Hyper-tune with seasonality component and train with full data """
    model_parameters = pd.DataFrame(columns = ['RMSE_value','Parameters'])
    for p in grid:
        m1 =Prophet(changepoint_prior_scale = p['changepoint_prior_scale'],
                    seasonality_prior_scale = p['seasonality_prior_scale'],
                    n_changepoints = p['n_changepoints'],
                    interval_width = p['interval_width'],
                    holidays_prior_scale=0,
                    yearly_seasonality = False,
                    seasonality_mode = 'additive'
                   )
        # Adds add_seasonality with standard params;
        # This should affect what values we get for the params_grid
        m1.add_seasonality(name='monthly', 
                  period= 30.5, 
                  fourier_order = 5,
                  prior_scale = 0.02
                 )
        # ok to use full data, to get optimal params        
        m1.fit(combined_partial)
        future1 = m1.make_future_dataframe(periods=8, freq='M')
        forecast1 = m1.predict(future1)[:592]
        RMSE_value = rmse(combined_partial['y'],abs(forecast1['yhat']))
        # populate model_parameters df
        model_parameters = model_parameters.append({'RMSE_value':RMSE_value,'Parameters':p},ignore_index=True)
        
    parameters = model_parameters.sort_values(by=['RMSE_value'])
    parameters = parameters.reset_index(drop=True)
    # Shows the df.head sorted with the lowest RMSE values in the top row
    print(parameters.head(3))
    print('\n')
    print(parameters['Parameters'][0])    
```

```python
%%time
get_model_params()
```

```python
# Use the optimal values from parametergrid to run another fbProphet model
# I found that using specifying add_seasonality options improves the model in this case

m = Prophet(
changepoint_prior_scale= 1,
n_changepoints= 25,
seasonality_mode='additive',
seasonality_prior_scale = 10,
holidays_prior_scale = 0,
yearly_seasonality = False,    
interval_width = 0.85)

m.add_seasonality(name='monthly', 
          period= 30.5, 
          fourier_order = 5,
          prior_scale = 0.02)

m.fit(train)
```

```python
forecast = m.predict(future)

predictions_24 = forecast.iloc[-24:]['yhat']
rmse24 = rmse(predictions_24,test['y'])

predictions_592 = forecast['yhat'][:592]
rmse592 = rmse(predictions_592,combined_partial['y'])

print(f'Hyper-tunned fbProphet RMSE for past 2 years: {rmse24}')
print(f'Hyper-tunned fbProphet RMSE for 50 years: {rmse592}')
```

```python
%%time
df_cv = cross_validation(m, initial= initial, period= period, horizon = horizon)
temp = performance_metrics(df_cv)
temp.rmse.mean()
```

```python
ax = forecast.plot(x='ds',y='yhat',label='Predictions',legend=True,figsize=(12,8),title='fbProphet-hyper-tunned')
combined_partial.plot(x='ds',y='y',label='Actual data',legend=True,ax=ax)
```
