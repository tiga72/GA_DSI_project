### Problem statement

Gold serves several functions in the world economy, and its link with financial and macroeco-
nomic variables are well-established (Pierdzioch et al., 2014a,b). It has a monetary value and
is sought after by central banks to be part of their international reserves which fulfil many
purposes (Gupta et al., 2014). It has industrial uses and can be transformed into jewellery. In
modern finance, it is used as a hedge against inflation and as a safe haven during crises.

It is clear that given the significance of gold in the modern world, the ability to provide
accurate forecasts into the future price of gold will be of utmost importance.

<a id='adf'></a>
### Augmented Dickeyâ€“Fuller test

Augmented Dickeyâ€“Fuller tests for Stationarity in a time series data<br>

#### Stationarity
Time series data is said to be stationary if it does not exhibit trends or seasonality. <br>
That is, the mean, variance and covariance is constant over time<br>

```python
def adf_test(series):
    """Pass in a time series and an optional title, returns an ADF report"""
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)
    for key,val in result[4].items():
        out[f'critical value ({key})']=val
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis. Data has a unit root and is non-stationary")
```

```python
# The ADF test is optional, as running auto_arima generates the
# The ADF test statistic -0.050597 is greater than critical value(5%)-2.866536
# So we fail to reject null hypothesis
adf_test(combined_s592['spot_gold'])
```

#### Split the data into train/test sets

A general rule: set the length of your test set equal(or more) to your intended forecast size.

- `test size greater than forecast size: OK`
- test size less than forecast size:    Not recommended

```python
# We cannot shuffle or use random samples of data in order to preserve chronological order in time series analysis
# So sklearn train_test_split() wont work here
train = combined_s592[:568]
test = combined_s592[568:]
```

```python
# Like GridSearchCV, auto arima helps to find the best parameters for the ARIMA model
# Notice here, it is absolutely fine to use all data as we looking for the best weights for the model
# Seasonal set to true and since this is monthly, we set m =12
# Note that the recommended model may not be the one with the closest fit. 
# The AIC score takes complexity into account, and tries to identify the best forecasting model.

aa = auto_arima(combined_s592['spot_gold'], seasonal=True,m=12, n_jobs= -1)
```

```python
# auto_arima gives basic arima: p,d,q as (2, 1, 2) and
# Seaonal component values: (0, 0, 0, 12)

aa_params = aa.get_params
print(f'auto_arima params: {aa_params}')
```

### Re-affirms the p,d,q values



auto_arima generates p,d,q values, we can use cross check these values using:<br>

<strong>Pandas diff()</strong> method to check 'd' values<br>
A <strong>PACF Plot</strong> can reveal recommended AR(p) orders, and an <strong>ACF Plot</strong> can do the same for MA(q) orders.<br>

NB: for both acf and pacf plots:<br>
Markers outside 'blue band' indicates that it is statistically significant

```python
# First Order Differencing
# Non-stationary data can be made to look stationary through differencing. 
# It calculates the difference between consecutive observations.
# auto_arima gives 'd' as 1, so we perform k_diff=1 (pandas differencing)
# The data should be stationary after k_diff=1 differencing and it is
df1 = diff(combined_s592['spot_gold'],k_diff=1)
adf_test(df1)
```

---
#### 'q'(MA) is equal to the first lag where the ACF value is above the significance level.

source: https://www.datasciencecentral.com/profiles/blogs/tutorial-forecasting-with-seasonal-arima

```python
# S is equal to the ACF lag with the highest value (typically at a high lag) (where (p,d,q)(P,D,Q)[S])
# Points above the blue band(95% confidence level), is a candidate value for MA
# From this plot: MA can be anywhere 1 to 45
# Parsimonious(simpler) models preferred to prevent overfitting (choose lower lags)
plot_acf(combined_s592['spot_gold'],lags =40);
```

---
#### 'p'(AR) is equal to the `first` lag where the PACF value is above the significance level.

```python
# AR value should be 1 to 2 (2 is just over the blue band)
plot_pacf(combined_s592['spot_gold'],lags =40);
```

This tells us that the `AR component should be more important than MA`. From the <a href='https://people.duke.edu/~rnau/411arim3.htm'>Duke University Statistical Forecasting site</a>:<br>
> <em>If the PACF displays a sharp cutoff while the ACF decays more slowly (i.e., has significant spikes at higher lags), we    say that the stationarized series displays an "AR signature," meaning that the autocorrelation pattern can be explained more    easily by adding AR terms than by adding MA terms.</em><br>

```python
# Basically a SARIMA model here, without the exogenous variable stated
# Uses only train data to train the model
model = SARIMAX(train['spot_gold'],order=(2, 1, 2),seasonal_order=(0, 0, 0, 12),enforce_invertibility=False)
results = model.fit()

# Looking at the column 'P>|Z|'(for ar.L and ma.L), all the p-values are less than 0.05, so the AR and MA values are ok
results.summary()
```

```python
start = len(train)
end =  len(train) + len(test) - 1
```

```python
predictions_2yrs = results.predict(start=start,end=end).rename('SARIMA Predictions_2yrs')
predictions_50yrs = results.predict().rename('SARIMA Predictions_50yrs')
```

```python
#predictions = results.predict().rename('SARIMA Predictions')
#gold['spot gold']['2010':].plot(legend=True, figsize=(16,5))
#predictions['2010':].plot(legend=True)
```

### Evaluating forecast accuracy
The choice of loss function:<br>
A forecast method that minimizes the MAE will lead to forecasts of the median, <br>
while minimizing the RMSE will lead to forecasts of the mean.

```python
# Akaike information criterion (AIC). Lower AIC value indicates better fit
error_may18_to_apr20 = rmse(test['spot_gold'],predictions_2yrs)
error_1971_to_apr20 = rmse(combined_s592['spot_gold'][:568],predictions_50yrs)
aic_sarima = aa.aic()

print(f'SARIMA RMSE for past 2 years: {error_may18_to_apr20}')
print(f'SARIMA model RMSE for 50 years: {error_1971_to_apr20}')
print(f'SARIMA AIC: {aic_sarima}')
```

<a id='gct'></a>
### Granger Causality Test

In building a SARIMAX model with exogenous variables<br>
But do we use some or all of the variables we have?<br>
Is there a direct and quantitative way to check causality factor?<br>
of a variable we are looking to add into our model??<br>

Granger causality test addresses the questions above and investigates causality between two variables in a time series<br>
p values lower than 0.05, indicates that the mentioned x(t) variable `granger causes` the target y(t), which is gold price

```python
def get_best_gct(gct_output_var, exog_var_name):
    """Returns no. of lags where the p value is lowest"""
    list1=[]
    for nlag in range(1,4,1): # maxlag=3 is standard
        temp_p_value = gct_output_var[nlag][0]['ssr_ftest'][1]
        list1.append(temp_p_value)
    min1 = min(list1) # gets the min value inside list
    min_posn = list1.index(min(list1)) + 1 # Need to plus 1 due to list item length
    print(F'{exog_var_name} minimum P value is {min1} with no.of lag at {min_posn}')
```

```python
# Running granger causality tests on the 3 potential exogenous variables
gct_result_cpi = grangercausalitytests(combined_s592[['spot_gold','us_cpi_inflation']],maxlag=3,verbose=False);
gct_result_rate = grangercausalitytests(combined_s592[['spot_gold','fed_interest_rate']],maxlag=3,verbose=False);
gct_result_usd = grangercausalitytests(combined_s592[['spot_gold','usd_index']],maxlag=3,verbose=False);
```

```python
# Runs the pre-defined function to check for lowest p-value and lag values
# The fed_interest_rate appears to be most significant with p-value at 0.04
get_best_gct(gct_result_cpi,'CPI')
get_best_gct(gct_result_rate,'Rates')
get_best_gct(gct_result_usd,'USD')
```

<a id='sx1'></a>
### SARIMAX models (with single exogenous variables)

Previously, I ran the Granger Causality Test with 'Rates' having statistically significant outcomes and 'USD' faring poorest

Without jumping into modelling 3 exogenous variables all at once <br>
I wanted to look at the performance when I ran 3 diifferent models, each time with just 1 exogenous variable.<br>

#### SARIMAX (Modelling Sequence)

<img src="../misc/sarimax_describe.jpg" width="650" align="left">

```python
# Again, can use the full data while getting best params from auto_arima
aa_for_cpi = auto_arima(combined_s592['spot_gold'], 
                        exogenous=combined_s592[['us_cpi_inflation']],
                        seasonal=True,
                        n_jobs= -1,
                        max_order = None,
                        m=12)
aa_for_rates = auto_arima(combined_s592['spot_gold'], 
                          exogenous=combined_s592[['fed_interest_rate']],
                          seasonal=True,
                          n_jobs= -1,
                          max_order = None,
                          m=12)
aa_for_usd = auto_arima(combined_s592['spot_gold'], 
                        exogenous=combined_s592[['usd_index']],
                        seasonal=True,
                        n_jobs= -1,
                        max_order = None,
                        m=12)

print(f'auto_arima params for cpi: {aa_for_cpi.get_params}')
print(f'auto_arima params for rates: {aa_for_rates.get_params}')
print(f'auto_arima params for usd: {aa_for_usd.get_params}')
```

```python
# Creating and fitting the models
cpi_model = SARIMAX(train['spot_gold'],
                    exog=train[['us_cpi_inflation']],
                    order=(2, 1, 2),
                    seasonal_order=(1, 0, 1, 12),
                    enforce_invertibility=False)
rates_model = SARIMAX(train['spot_gold'],
                      exog=train[['fed_interest_rate']],
                      order=(2, 1, 2),
                      seasonal_order=(0, 0, 0, 12),
                      enforce_invertibility=False)
usd_model = SARIMAX(train['spot_gold'],
                    exog=train[['usd_index']],
                    order=(2, 1, 2),
                    seasonal_order=(0, 0, 0, 12),
                    enforce_invertibility=False)

cpi_result = cpi_model.fit()
rates_result = rates_model.fit()
usd_result = usd_model.fit()
```

```python
# Generates predictions to be used for RMSE caculations
cpi_predictions = cpi_result.predict(start=start,end=end, exog=test[['us_cpi_inflation']]).rename('SARIMAX1 Predictions')
rates_predictions = rates_result.predict(start=start,end=end, exog=test[['fed_interest_rate']]).rename('SARIMAX2 Predictions')
usd_predictions = usd_result.predict(start=start,end=end, exog=test[['usd_index']]).rename('SARIMAX3 Predictions')
```

```python
# Evaluates all 3 individual models base on the loss function and AIC values(lower better)
# RMSE errors outcome coincides with Granger Causality Test with 'rates' having best and 'USD having worst values
# AIC 

cpi_error = rmse(test['spot_gold'],cpi_predictions)
rates_error = rmse(test['spot_gold'],rates_predictions)
usd_error = rmse(test['spot_gold'],usd_predictions)

cpi_aic = aa_for_cpi.aic()
rates_aic = aa_for_rates.aic()
usd_aic = aa_for_usd.aic()

print(f'RMSE for CPI: {cpi_error}')
print(f'RMSE for Rates: {rates_error}')
print(f'RMSE for USD: {usd_error}'+'\n')

print(f'AIC for CPI: {cpi_aic}')
print(f'AIC for Rates: {rates_aic}')
print(f'AIC for USD: {usd_aic}')
```

```python
sx3b_model = SARIMAX(combined_s592['spot_gold'],
                        exog=combined_s592[['us_cpi_inflation','fed_interest_rate','usd_index']],
                        order=(2, 1, 3),
                        seasonal_order=(0, 0, 0, 12),
                        enforce_invertibility=False)

sx3b_results = sx3b_model.fit()
```

```python
exog_fcast3 = combined_s600[592:][['us_cpi_inflation','fed_interest_rate','usd_index']]

fcast_24mths = sx3b_results.predict(start='2018-05-01',end='2020-04-01',exog=exog_fcast3).rename('fcast_24mths')

fcast_all_3var = sx3b_results.predict(exog=exog_fcast3).rename('fcast_all_3var')

sx3b_plot = sx3b_results.predict(start='1971-01-01',
                                   end='2020-12-01',
                                   exog=combined_s600[592:][['us_cpi_inflation','fed_interest_rate','usd_index']]).rename('sx3b_plot')
```

```python
error_may18_to_apr20 = rmse(test['spot_gold'],fcast_24mths)
error_1971_to_apr20 = rmse(combined_s592['spot_gold'],fcast_all_3var)
final_3var = sx3_aa.aic()

print(f'Full model RMSE for past 2 years: {error_may18_to_apr20}')
print(f'Full model RMSE for 50 years: {error_1971_to_apr20}')
print(f'Final AIC: {final_3var}')
```
